```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
```

```python
data.columns = [x.lower() for x in data.columns]
```

```python
data['комментарий'] = data['комментарий'].astype(str)
```

```python
data['оценка'].hist();
```

Большинство отзывов в диапозоне 9-10. Стоит рассмотреть основные боли по оценкам 0-6.

```python
data[data['оценка']<=6]
```

После просмотра данных сделаем функцию которая посчитатет основные слова - "боли"

```python
categories = ['долго', 'сложно', 'непонятно', 'проблемы', 'неправильно',
              'не помог', 'не нравится', 'несправедливо', 'строг', 'недопонимание']


# Инициализация столбцов для каждой категории
for category in categories:
    data[category] = data['комментарий'].str.contains(category, case=False, na=False).astype(int)


category_counts = data[categories].sum().sort_values()

# Визуализация данных
category_counts.plot(kind='bar')
plt.title('Частота упоминания основных проблем в комментариях учеников')
plt.xlabel('Категории')
plt.ylabel('Количество упоминаний')
plt.show()


category_counts
```

Чаще всего встречаются такие слова как: долго, строг, не нравится, непонятно.

```python
# Создание категорий
categories_2 = ['быстро', 'понятно', 'добрый', 'легко', 'нравится', 'помог', 'справедливо', 'умный']


# Инициализация столбцов для каждой категории
for category in categories_2:
    data[category] = data['комментарий'].str.contains(category, case=False, na=False).astype(int)


category_counts_2 = data[categories_2].sum().sort_values()

# Визуализация данных
category_counts_2.plot(kind='bar')
plt.title('Частота упоминания основных проблем в комментариях учеников')
plt.xlabel('Категории')
plt.ylabel('Количество упоминаний')
plt.show()


category_counts_2
```

В обратном потсчете: быстро, понятно, нравится, помог
